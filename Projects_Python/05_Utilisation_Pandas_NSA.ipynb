## Filtrer Une Base De Données En Python

```python
# Créer une base de données des femmes --> Filtre sur le sexe F
df_femmes = df_titanic[df_titanic['Sex'] == "female"]
df_femmes
```

```python
df_femmes.Sex.unique()
```

```python
df_titanic.Sex.unique()
```

```python
# Créer une base de données des femmes ayant survécu --> Filtre sur le sexe F et Survived
df_femmes_survived = df_titanic[(df_titanic['Sex'] == "female") & (df_titanic['Survived'] == 1)]
df_femmes_survived
```

```python
df_femmes_survived.Survived.unique()
```

```python
# Créer une base de données des hommes n'ayant pas survécu --> Filtre sur le sexe 'male' et 'Survived' == 0
df_hommes_not_survived = df_titanic[(df_titanic['Sex'] == "male") & (df_titanic['Survived'] == 0)]
df_hommes_not_survived
```

```python
# Créer une base de données des personnes mineures dans le titanic --> Filtre sur 'Age' < 21
df_mineure = df_titanic[df_titanic.Age < 18]
df_mineure
```

```python
# moyenne d'age des mineurs
df_mineure.Age.mean()
```

```python
# Age médian des mineurs
df_mineure.Age.median()
```

## Mettre à Jour Un Data Frame (Update)

```python
# Update
# Créer une variable qui dit si on est mineur
df_titanic["Est_Mineur"] = df_titanic["Age"] < 18
```

```python
df_titanic.head()
```

```python
# Moyenne d'âge des personnes qui ne sont pas mineures
temp = df_titanic[df_titanic["Est_Mineur"] == False ] 
temp.Age.mean()
```

```python
# Update
# Rajout d'une colonne : age au carré
df_titanic["Age_carre"] = df_titanic["Age"] ** 2
df_titanic
```

```python
df_titanic.info()
```

```python
# Update
# Suppression de colonnes : colonne "Est_Mineur"
df_titanic.drop(["Est_Mineur"], axis = 1) # axis = 0 (suppression par ligne) ; axis = 1 (suppression par colonne)
```

```python
df_titanic
```

```python
# Pour prendre en compte la suppression dans la base de données il faut créer un nouvelle BB ou utiliser la même
df_titanic = df_titanic.drop(["Est_Mineur"], axis = 1)
```

```python
df_titanic
```

```python
# Update
# Suppression de colonnes : colonne "Age_carre"
df_titanic = df_titanic.drop(["Age_carre"], axis = 1)
df_titanic
```

```python
# Update
# Suppression de colonnes à ne pas utiliser dans le modèle : colonne "Ticket"
df_titanic = df_titanic.drop(["Ticket"], axis = 1)
df_titanic
```

## Différence entre .loc et .iloc

```python
# .iloc : permet de récupérer des informations dans les DF à partir de l'index
df_titanic.iloc[0] # info sur la 1ère lignes
```

```python
df_titanic.iloc[0:3] # info sur les 3 premières lignes
```

```python
# info sur les 3 premières lignes et de la 2ème à la 4ème colonne
df_titanic.iloc[0:3, 1:4]
```

```python
# .loc : permet de récupérer des informations dans les DF à partir de l'index, 
# par contre à la différence de .iloc, lorsqu'il s'agit des colonnes il faut rajouter les noms de colonnes
df_titanic.loc[0:3, 1:4]
```

```python
df_titanic.loc[0:3]
```

```python
df_titanic.loc[0:3, ["Survived", "Pclass", "Sex",]]
```

## Traitement Des Données Manquantes

```python
df_titanic.isnull().sum()
```

```python
# astuce, lorqu'une variable a plus de 15% de données manquantes je n'utilise pas cette variable dans mon modèle
```

```python
# remplace les valeurs manquantes de Age par la moyenne d'age
df_titanic['Age'].fillna(df_titanic['Age'].mean(), inplace = True) 
```

```python
df_titanic['Age'].mean()
```

```python
df_titanic.isnull().sum()
# on n'a plus de valeurs manquante sur Age
```

```python
df_titanic.dropna(subset = ["Embarked"], inplace = True)
# df_titanic = df_titanic.dropna(subset = ["Embarked"]) --> même résultat que la ligne précédente
```

```python
df_titanic.isnull().sum()
# on n'a plus de valeurs manquante sur Embarked
```

```python
df_titanic.info()
```

```python
df_titanic.isnull().mean()
#df_titanic.isna().mean() --> isna() est équivalent à isnull()
```
